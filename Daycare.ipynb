#### <font color='blue'>Import packages and load dataset.</font>

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
filename = 'D:/Statistics/640/Term Project/nationaldatabaseofchildcareprices.xlsx'
dataframe = pd.read_excel(filename)

dataframe['State_Abbreviation']=dataframe['State_Abbreviation'].astype('category')
dataframe['state_codes'] = dataframe['State_Abbreviation'].cat.codes
```

#### <font color='blue'>Create six new datasets for daycare prices and incomes.</font>

```python
# get daycare prices for each age group
infant = dataframe.groupby('StudyYear')['MCInfant'].mean().reset_index()
toddler = dataframe.groupby('StudyYear')['MCToddler'].mean().reset_index()
preschool = dataframe.groupby('StudyYear')['MCPreschool'].mean().reset_index()

# get incomes from dataset
income = dataframe.groupby('StudyYear')['MHI'].mean().reset_index()
fme = dataframe.groupby('StudyYear')['FME'].mean().reset_index()
mme = dataframe.groupby('StudyYear')['MME'].mean().reset_index()
```

#### <font color='blue'>Create box plots of daycare prices to show spread of data and outliers.</font>

```python
# box plots

dataframe.boxplot(column=['MCInfant','MCToddler','MCPreschool'])
plt.title("Daycare Cost Box Plots")
plt.xlabel('Age Group')
plt.ylabel('Cost in Dollars')
plt.show() 
```

```python
fig = plt.figure(figsize = (10, 5))

infant_loc = dataframe.groupby('State_Abbreviation')['MCInfant'].mean().reset_index()
infant_loc = infant_loc.query('MCInfant > 250')
#toddler = dataframe.groupby('StudyYear')['MCToddler'].mean().reset_index()
#preschool = dataframe.groupby('StudyYear')['MCPreschool'].mean().reset_index()

# creating the bar plot
plt.bar(infant_loc['State_Abbreviation'], infant_loc['MCInfant'], color ='blue', 
        width = 0.4)
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.xlabel("Location")
plt.ylabel("Cost in Dollars")
plt.title("Infant Childcare Cost by Area")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))

toddler_loc = dataframe.groupby('State_Abbreviation')['MCToddler'].mean().reset_index()
toddler_loc = toddler_loc.query('MCToddler > 220')
#toddler = dataframe.groupby('StudyYear')['MCToddler'].mean().reset_index()
#preschool = dataframe.groupby('StudyYear')['MCPreschool'].mean().reset_index()

# creating the bar plot
plt.bar(toddler_loc['State_Abbreviation'], toddler_loc['MCToddler'], color ='blue', 
        width = 0.4)
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.xlabel("Loation")
plt.ylabel("Cost in Dollars")
plt.title("Toddler Childcare Cost by Area")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))

preschool_loc = dataframe.groupby('State_Abbreviation')['MCPreschool'].mean().reset_index()
preschool_loc = preschool_loc.query('MCPreschool > 200')
#toddler = dataframe.groupby('StudyYear')['MCToddler'].mean().reset_index()
#preschool = dataframe.groupby('StudyYear')['MCPreschool'].mean().reset_index()

# creating the bar plot
plt.bar(preschool_loc['State_Abbreviation'], preschool_loc['MCPreschool'], color ='blue', 
        width = 0.4)
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.xlabel("Location")
plt.ylabel("Cost in Dollars")
plt.title("Preschool Childcare Cost by Area")
plt.show()
```

#### <font color='blue'>Create line charts of daycare prices and incomes.</font>

```python
# multiply each by 48 to get yearly cost
plt.plot(infant['StudyYear'], infant['MCInfant']*48, color='#004cff')
plt.plot(toddler['StudyYear'], toddler['MCToddler']*48, color='#EE822B')
plt.plot(preschool['StudyYear'], preschool['MCPreschool']*48, color='#2BEE4B')
plt.legend(['Infant', 'Toddler', 'Preschool'])
plt.title("Daycare Cost")
plt.xlabel('Date')
plt.ylabel('Yearly Cost in Dollars')
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.show()

plt.plot(income['StudyYear'], income['MHI'], color='#000000')
plt.legend(['Median Income'])
plt.title("Median Income")
plt.xlabel('Date')
plt.ylabel('Yearly Income in Dollars')
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.show()

plt.plot(fme['StudyYear'], fme['FME'], color='#FF00FF')
plt.plot(mme['StudyYear'], mme['MME'], color='#004cff')
plt.legend(['Female Income', 'Male Income'])
plt.title("Income by Sex")
plt.xlabel('Date')
plt.ylabel('Yearly Income in Dollars')

for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
    
plt.show()
```

#### <font color='blue'>Show birth rates by age groups of mothers.</font>

```python
# load dataset
filename = 'D:/Statistics/640/Term Project/wages_by_education.csv'
df = pd.read_csv(filename)

# calculate yearly incomes
df.loc[:,'high_school'] *= 1920
df.loc[:,'some_college'] *= 1920
df.loc[:,'bachelors_degree'] *= 1920
df.loc[:,'advanced_degree'] *= 1920

# filter out each level
hs = df.groupby('year')['high_school'].mean().reset_index()
sc = df.groupby('year')['some_college'].mean().reset_index()
bd = df.groupby('year')['bachelors_degree'].mean().reset_index()
ad = df.groupby('year')['advanced_degree'].mean().reset_index()
```

```python
# plot lines of different education level salaries
plt.plot(hs['year'], hs['high_school'], color='#004cff')
plt.plot(sc['year'], sc['some_college'], color='#EE822B')
plt.plot(bd['year'], bd['bachelors_degree'], color='#2BEE4B')
plt.plot(ad['year'], ad['advanced_degree'], color='#eb344c')
plt.legend(['High School', 'Some College', 'Bachelors', 'Advanced'])
plt.title("Salary by Education Level")
plt.xlabel('Date')
plt.ylabel('Salary')

#remove top and right border
for pos in ['right', 'top']: 
    plt.gca().spines[pos].set_visible(False) 
plt.show()
```
